```python
import  matplotlib.pyplot as plt
f = open('day-3.txt')
instr1 = f.readline()
instr2 = f.readline()
```

```python
data1 = instr1.split(',')
data2 = instr2.split(',')
data = ['U7','R6','D4','L4']
```

```python

def getPath(data):

    rhist = [0,]
    chist = [0,]
    totdist = [0,]

    r = 0
    c = 0

    for item in data:
        #print(item)
        way = item[0]
        dist = int(item[1:])

        if way=='R':
            c += dist
        elif way=='L':
            c -= dist
        elif way=='U':
            r += dist
        elif way=='D':
            r -= dist
        else:
            raise('ERROR')
        rhist.append(r)
        chist.append(c)
        totdist.append(totdist[-1] + dist)
    return rhist,chist, totdist
    #print(way,dist)
```

```python
r1,c1,dist1 = getPath(data1)
r2,c2,dist2 = getPath(data2)
```

```python
%matplotlib notebook
plt.figure(figsize=(8,6))
plt.plot(r1,c1)
plt.plot(r2,c2)


plt.figure(figsize=(8,6))
plt.plot(r1,c1)
plt.plot(r2,c2)

plt.xlim(-400,400)
plt.ylim(-400,400)

#print(829+307)
```
